```python
import pandas as pd
from sklearn import datasets

iris = datasets.load_iris()

```

```python
import matplotlib.pyplot as plt

_, ax = plt.subplots()
scatter = ax.scatter(iris.data[:, 0], iris.data[:, 1], c=iris.target)
ax.set(xlabel=iris.feature_names[0], ylabel=iris.feature_names[1])
_ = ax.legend(
    scatter.legend_elements()[0], iris.target_names, loc="lower right", title="Classes"
)

_, ax = plt.subplots()
scatter = ax.scatter(iris.data[:, 2], iris.data[:, 3], c=iris.target)
ax.set(xlabel=iris.feature_names[2], ylabel=iris.feature_names[3])
_ = ax.legend(
    scatter.legend_elements()[0], iris.target_names, loc="lower right", title="Classes"
)
```

```python
X_full , y_full = iris.data , iris. target
feature_names = iris.feature_names
# Access the target variable (data labels)
target = iris. target
target_names = iris.target_names
# Print the name of the target variable
print( "target_names" )
print(target_names)
# Print the feature names
print("Feature Names: " )
print(feature_names)

df = pd.DataFrame(X_full, columns=feature_names)

df.head()
print("df")
print(df)
```

```python
# Select the desired features
features = ["sepal length (cm)", "sepal width (cm)"]
features_idx = [list(feature_names).index(feature) for feature in features]
X = X_full[:, features_idx]

# Create a pandas DataFrame with the selected features
df = pd.DataFrame(X, columns=features)


# Calculate statistics for each feature using the describe() function
statistics = df.describe()

# Display the statistics for each feature
print(statistics)
```

```python
from sklearn.preprocessing import MinMaxScaler

scalerM = MinMaxScaler()

X_scaled=scalerM.fit_transform(X)




```

```python
plt.subplot(1, 2, 1)
plt.scatter(df["sepal length (cm)"], df["sepal width (cm)"])
plt.xlabel("sepal length (cm)")
plt.ylabel("sepal width (cm)")
plt.title("Data Before Normalization")

# Plot the data after normalization
plt.subplot(1, 2, 2)
plt.scatter(X_scaled[:, 0], X_scaled[:, 1])
plt.xlabel("Normalized sepal length (cm)")
plt.ylabel("Normalized sepal width (cm)")
plt.title("Data After Min-Max Scaling")

plt.tight_layout()
plt.show()
```
